## 3.1. Single-variable Linear Regression

We will first **carry out single-variable regressions for each variable** that might influence the percentage of women inventors across different countries within the same time period (2010-2014). The goal of this analysis is to figure out which variables should be included in our next analysis, the multivariable linear regression model building.

Predicting variables:

- Percentage of women declaring that they have access to training on how to start a business

- Percentage of women preferring taking a risk and starting a new business to working for someone else
    
Therefore, there will be a total of five linear regressions. We will calculate the **correlation coefficient** and **R square (coefficient of determination)** for each linear regression to examine whether there is a strong linear correlation between these variables and our dependent variable, which is the percentage of women inventors.

We will also calculate the **p-values** for each regression to check their significance, together with a **hypothesis test**.

With the results from single-variable regressions, we can then figure out which variables to include in our next step, which is the multi-variable linear regression. To ensure a reliable multi-variable linear regression, we will **exclude the variables that have the highest correlation and r square value** we will find in the single-variable linear regression models.

## 3.2. Multivariable Linear Regression

In reality, many powerful linear regression models are not limited to single-input variables, given that we are living in a complex world with countless variables. **We will then carry out multivariable regressions** for those selected variables that might influence the percentage of women inventors across different countries within the same time period (2010-2014). For those highly correlated variables and might have a causal effect on each other, **we will exclude the ones with the highest correlation and r square value**. For example, “Percentage of women declaring that they have access to training on how to start a business” and “Percentage of women declaring that they have access to money to start a business” might be highly correlated, and adding both variables into the multivariable regression might not bring about much new information compared to introducing only one of them.

We will use a **permutation test** to evaluate the significance of the multivariable regression, which is equivalent to a hypothesis test, but with better generalization to our real-world application given the limited size of our sample.

We can **compare the r squares for the single regressions and multivariable regression** to see which has a higher coefficient of determination. Therefore we are able to predict the relationship between the variables more accurately.

With the comparison of results from both single and multivariable regression models, we can understand what aspect OECD countries need to pay attention to and what kind of policies they need to make if these countries want to improve the share of women inventors.

### 4.2.1. Single-variable Linear Regression

Predicting variables:
- Gross enrolment ratio, tertiary, gender parity index (GPI)
- Percentage of graduates from Science and Technology programmes in tertiary education who are female
- Percentage of women declaring that they have access to training on how to start a business
- Percentage of women declaring that they have access to money to start a business
- Percentage of women preferring taking a risk and starting a new business to working for someone else

We will build 5 single-variable linear models to predict 'Share of Women Inventors (%)' from the variables above.

Then, we will calculate the **correlation coefficient** and **R square (coefficient of determination)** for each linear regression to examine whether the correlation is strong.

We will also calculate the **p-values** for each regression to check their significance, together with a **hypothesis test**.

With the results from single-variable regressions, we can then figure out which variables to include in our next step, which is the multivariable linear regression. To ensure a reliable multivariable linear regression, we will **exclude the variables that have the highest correlation and r square value**

```python
# We will construct a function "lr_helper"
# to help us with single-variable linear regression
# model building and graph plotting

def lr_helper(col_name, x_label, graph_title):
    gpi_dropna = final.dropna(subset=['Country', col_name])
    grouped_gpi = gpi_dropna.groupby('Country')
    
    X = grouped_gpi[col_name].mean()
    Y = grouped_gpi['Share of Women Inventors (%)'].mean()
    
    regressor = LinearRegression()
    regressor.fit(X.values.reshape(-1,1), Y)
    
    sns.regplot(x=X, y=Y)
    plt.xlabel(x_label)
    plt.ylabel('Share of Women Inventors (%)')
    plt.title(graph_title, size=20)
    plt.show()
    
    print('Predictor: ' + col_name)
    print('Slope:', regressor.coef_)
    print('Intercept:', np.round(regressor.intercept_, decimals=4))
    print('Correlation Coefficient:', np.round(np.corrcoef(X,Y)[0,1], decimals=4))
    print('R square (coefficient of determination):', np.round(regressor.score(X.values.reshape(-1,1), Y), decimals=4))
```

```python
# Linear Regression: Female Inventors (%) vs GPI

lr_helper(col_name = preds[0], 
          x_label = 'GPI', 
          graph_title = 'Female Inventors (%) vs GPI [LinearRegression]')


# Linear Regression: Female Inventors (%) vs Female Science-Tech Graduates (%)

lr_helper(col_name = preds[1], 
          x_label = 'Percentage of Female Science-Tech Graduates (%)', 
          graph_title = 'Female Inventors (%) vs Female Science-Tech Graduates (%) [LinearRegression]')


# Linear Regression: Female Inventors (%) vs Access to Training (%)

lr_helper(col_name = preds[2], 
          x_label = 'Have Access to Training (%)', 
          graph_title = 'Female Inventors (%) vs Access to Training (%) [LinearRegression]')


# Linear Regression: Female Inventors (%) vs Access to Money (%)

lr_helper(col_name = preds[3], 
          x_label = 'Have Access to Money (%)', 
          graph_title = 'Female Inventors (%) vs Access to Money (%) [LinearRegression]')


# Linear Regression: Female Inventors (%) vs Positive Attitude (%)

lr_helper(col_name = preds[4], 
          x_label = 'Positive Attitude (%)', 
          graph_title = 'Female Inventors (%) vs Positive Attitude to Risks (%) [LinearRegression]')
```

### 4.2.2. Multivariable Linear Regression

**We will then carry out multivariable regressions** for those selected variables that might influence the percentage of women inventors across different countries within the same time period (2010-2014). For those predicting variables that are highly correlated and might have a causal effect on each other, **we will exclude one of them**.

After we obtained a multivariable regression model, we will calculate its **correlation coefficient** and **r square score** to examine whether our multivariable regression has a strong explanatory power (i.e., explaining the percentage of women inventors in countries that are included in our model).

We will also calculate the **p-value** for the multivariable regression model to check its significance using a **permutation test**.

We can **compare the r squares for the single regressions and multivariable regression** to see which has a higher coefficient of determination, and therefore we are able to predict the relationship between the variables more accurately.

```python
# we will construct a helper function "multi_helper"
# to help us with multivariable linear regression
# model building

def multi_helper(variables):
    X = grouped_dropna[variables]
    Y = grouped_dropna['Share of Women Inventors (%)']

    multi_model = LinearRegression()
    multi_model.fit(X, Y)

    multi_slope = multi_model.coef_
    multi_r2 = multi_model.score(X,Y)
    
    print( "Coefficient for GPI:                      {:.2f}".format(multi_slope[0]))
    print( "Coefficient for Female Sci-Tech Graduates:{:.2f}".format(multi_slope[1]))
    if preds[3] not in variables:
        print( "Coefficient for Access to Training:       {:.2f}".format(multi_slope[2]))
        print( "Coefficient for Positive Attitude:        {:.2f}".format(multi_slope[3]))
    elif preds[2] not in variables:
        print( "Coefficient for Access to Money:          {:.2f}".format(multi_slope[2]))
        print( "Coefficient for Positive Attitude:        {:.2f}".format(multi_slope[3]))
    else: 
        print( "Coefficient for Access to Money:          {:.2f}".format(multi_slope[2]))
        print( "Coefficient for Access to Money:          {:.2f}".format(multi_slope[3]))
        print( "Coefficient for Positive Attitude:        {:.2f}".format(multi_slope[4]))
    print()
    print( "r^2 score of the multivariable model:     {:.3f}".format(multi_r2))
```

```python
# We will construct a multivariable linear regression model
# with all five predictor variables.

preds = [
    'Gross enrolment ratio, tertiary, gender parity index (GPI)',
    'Percentage of graduates from Science and Technology programmes in tertiary education who are female (%)',
    'Percentage of women declaring that they have access to training on how to start a business (%)',
    'Percentage of women declaring that they have access to money to start a business (%)',
    'Percentage of women prefering taking a risk and starting a new business to working for someone else (%)']

multi_helper(preds)
```

```python
# We will construct a multivariable linear regression model
# with four predictor variables, excluding "percentage of ... access to training"

reduced_preds_1 = [
    'Gross enrolment ratio, tertiary, gender parity index (GPI)',
    'Percentage of graduates from Science and Technology programmes in tertiary education who are female (%)',
#     'Percentage of women declaring that they have access to training on how to start a business (%)',
    'Percentage of women declaring that they have access to money to start a business (%)',
    'Percentage of women prefering taking a risk and starting a new business to working for someone else (%)']

multi_helper(reduced_preds_1)
```

```python
# We will construct a multivariable linear regression model
# with four predictor variables, excluding "percentage of ... access to money"

reduced_preds_2 = [
    'Gross enrolment ratio, tertiary, gender parity index (GPI)',
    'Percentage of graduates from Science and Technology programmes in tertiary education who are female (%)',
    'Percentage of women declaring that they have access to training on how to start a business (%)',
#     'Percentage of women declaring that they have access to money to start a business (%)',
    'Percentage of women prefering taking a risk and starting a new business to working for someone else (%)']

multi_helper(reduced_preds_2)
```

How about using this model to predict the 'Share of Women Inventors (%)' for Australia, Chile, France, United Kingdom, and United States in 2014?

```python
# display the dataset piece

final.loc[ (final['Year'] == 2014)
          &
          ( (final['Country'] == 'Australia')
          | (final['Country'] == 'Chile')
          | (final['Country'] == 'France')
          | (final['Country'] == 'United Kingdom')
          | (final['Country'] == 'United States')
          )] 
```

```python
# we will construct a function "multi_result"
# to calculate the 'Share of Women Inventors (%)'
# based on our multivariable linear regression model

def multi_result(gpi, scitech, money, attitude):
    coef_gpi = 0.46
    coef_scitech = 0.13
    coef_money = -0.15
    coef_attitude = 0.15
    
    result = gpi * coef_gpi + scitech * coef_scitech + money * coef_money + attitude * coef_attitude
    return result
```

```python
def permute(input_array):
    # shuffle is inplace, so copy to preserve input
    permuted = input_array.copy()
    np.random.shuffle(permuted)
    return permuted
```

```python
# plot one instance of permuted data
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,6), sharey=True)
sns.regplot(x=X['Gross enrolment ratio, tertiary, gender parity index (GPI)'], y=Y, ax=ax1)
ax1.set_title('True data')
sns.regplot(x=X['Gross enrolment ratio, tertiary, gender parity index (GPI)'], y=permute(Y), ax=ax2)
ax2.set_title('Permuted data')
plt.show()
```

```python
def plot_model_line(df, model, is_resampled=True):
    '''
    Takes a dataframe and a fitted model
    Plots a line of best fit to the data
    '''
    if is_resampled:
        color="grey"
        alpha=0.1
    else:
        color="steelblue"
        alpha=0.7
    plt.plot(df['x'], df['x'] * model.coef_[0] + model.intercept_, color=color, alpha=alpha)
```

```python
# plot true data
plt.scatter(X['Gross enrolment ratio, tertiary, gender parity index (GPI)'], Y)

# plot 100 iterations of models on permuted data
for i in range(100):
    multi_model.fit(X, Y)
    fake_model = LinearRegression().fit(X, permute(Y))
    plot_model_line(df, fake_model)

# model true data and plot
model = LinearRegression().fit(X, Y)
plot_model_line(df, model, is_resampled=False)
    
plt.show()
```

```python
steeper_slopes = 0
n_permutations = 10000
permuted_slopes = np.zeros(n_permutations)

for i in range(n_permutations):
    fake_model = LinearRegression().fit(X, permute(Y))
    
    permuted_slopes[i] = fake_model.coef_[0]
    
    if np.abs(fake_model.coef_[0]) > np.abs(model.coef_[0]):
        steeper_slopes += 1
    
print(steeper_slopes / n_permutations)
```

```python
binsizes, _, _ = plt.hist(permuted_slopes, bins=50)
plt.vlines(multi_model.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.show()
```

```python
# compare simulated probability of observing a slope this great or greater
#  to analytic p-value
#print('Analytic p (slope): ', result.pvalues.x)
print('Simulated p (slope):', steeper_slopes / n_permutations)
```

From the permutation test above, we can see that **the p-value (steeper_slopes / n_permutations) is more than 0.9**, which is far larger than 0.05, and therefore our model may not be able to predict for a larger population in the real-world. The result is expected, since our dataset is **too small**, and our original data is scattered around with a **noisy weak relationship.** The permutation test helps us further verify this.

However, we are also aware that applying permutation test to multivariable regression might be tricky, as it is not as accurate as it is for single-varibale regression, given the complex nature with multiple variables. Therefore, we are also conservative about the result we obtained.

# 7. Limitations

One limit of our dataset is that we didn't include the public opinions on women inventors (or entrepreneurs), which can be an important externality that influence the share of women inventors. Though we thought about this critical factor, we didn't include it since public opinions are hard to be quantized. We didn't find suitable data source that include data from the countries and timeframe we are researching on either.

In terms of harm and gap, our dataset is also limited by the choice of countries. We only focus on OECD member countries, which are all already developed. Therefore, our dataset can only reflect the situation of women inventors in developed countries, but we are unable to explore that of developing and under-developed ones. The population of developed countries only make up of a small proportion of the world population, so our dataset cannot reflect the overall realistic situation of women. If we want to actually figure out underlying trend and reasons that influence share of women inventors, we should go beyond just looking at developed countries. We should in turn focus more on missing data and try to collect data from all over the world. However, due to the limited time and resource, we only focus on existing data on developed countries, hoping to get a preliminary understanding of the topic of women inventors. Another associating limitation due to choice of countries is that there are only 33 OECD member countries, making our dataset relatively small in the scale. 

Besides, our dataset is also limited in the timeframe we chose. We only focus on data from 2010-2014, which is only a 5-year period. This period of time is also over 5 year from now. Though we believe this is a representative and meaningful period of time, our dataset is relatively small in its scale. There may be some other trends and influential factors that might influence share of women inventors, but we cannot include them.

Last but not least, the multivariable linear regression model requires us to drop all NaN's in the original dataset, so this processing left us with only 26 OECD countries as the input dataset of our multivariable model. Though we have a time frame of five years, our dataset is still quite limited because we will have to take the data's mean in five years. Our multivariable regression obtained from this processed dataset might be over-fitting and do not apply to non-OECD countries; however, the research method and model building procedure is generalizable and definitely can be used to carry out study on non-OECD countries, if required data are collected.

In all, though we have included multiple valuable variables that may influence share of women inventors that can potentially bring interesting discussion and analysis of the issue, our dataset size is not massive enough to conclude a comprehensive result.
